```python
from bs4 import BeautifulSoup

import urllib
import re
```

```python
r = urllib.urlopen('https://en.wikipedia.org/wiki/Web_scraping').read()
soup = BeautifulSoup(r,'lxml')
type(soup)
```

```python
print soup.prettify()[:100]
```

```python
for link in soup.find_all('a'):
    print link.get('href')
```

```python
for link in soup.find_all('a', attrs = {'href': re.compile("http")}):
    print link
```

```python
## Saving in a file

file = open('ParsedText.txt','wb' )
for link in soup.find_all('a', attrs = {'href': re.compile("http")}):
    ## First convert to string
    link = str(link)
    file.write(link)
file.flush()
file.close()
```
